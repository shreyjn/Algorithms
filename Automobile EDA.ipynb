### Cleaning Price

```python
df.price.str.isnumeric().value_counts()
```

```python
df['price']= df['price'].apply(lambda x: 0 if x=='?' else int(x))
```

```python
x=df['price'].loc[df['price']!=0].values
```

```python
df['price']= df['price'].apply(lambda i: x.mean() if i==0 else i)
```

### Cleaning Bore

```python
df['bore'].loc[df['bore']=='?']
```

```python
df['bore']= df['bore'].apply(lambda x: 0 if x=='?' else float(x))
x=df['bore'].loc[df['bore']!=0].values
df['bore']= df['bore'].apply(lambda i: x.mean() if i==0 else i)
```

### Cleaning Horsepower

```python
df['horsepower'].loc[df['horsepower'].str.isnumeric()==False]
```

```python
df['horsepower']= df['horsepower'].apply(lambda x: 0 if x=='?' else float(x))
x=df['horsepower'].loc[df['horsepower']!=0].values
df['horsepower']= df['horsepower'].apply(lambda i: x.mean() if i==0 else i)
```

```python
df['num-of-cylinders'].value_counts()
```

### Cleaning Peak-rpm

```python
df['peak-rpm'].loc[df['peak-rpm'].str.isnumeric()==False]
```

```python
df['peak-rpm']= df['peak-rpm'].apply(lambda x: 0 if x=='?' else int(x))
x=df['peak-rpm'].loc[df['peak-rpm']!=0].values
df['peak-rpm']= df['peak-rpm'].apply(lambda i: x.mean() if i==0 else i)
```

### BiVariate Analysis

```python
plt.figure(figsize=(20 ,15))
sns.boxplot(x=df.make, y=df.price);
```

```python
plt.figure(figsize=(10 ,8))
sns.boxplot(x=df['fuel-type'], y=df['price']);
```

```python
plt.figure(figsize=(10 ,10))
sns.jointplot(df.price, df['engine-size'], color='r', kind='reg');
```

```python
plt.figure(figsize=(10, 10))
sns.boxplot(x=df['body-style'], y=df.price);
```

```python
plt.figure(figsize=(18 ,10))
df.groupby('body-style')['make'].value_counts().plot(kind='bar');
plt.ylabel("NO. of Cars");
```

```python
plt.figure(figsize=(10 ,10))
sns.jointplot(df.horsepower, df['peak-rpm'], kind='reg', color='darkorange');
```

```python
plt.figure(figsize=(10 ,10))
sns.jointplot(df['wheel-base'], df['peak-rpm'], kind='reg', color='darkorange');
```
